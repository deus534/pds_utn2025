```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000
fs = 10000

def mi_funcion_sen(vmax = 1, dc = 0, ff = 1, ph = 0, nn = N, fs = fs):
    t = nn/fs
    tt = np.arange(0,t,1/fs)
    xx = vmax*np.sin(2*np.pi*ff*tt + ph) + dc

    return tt, xx
```

```python
tt, xx = mi_funcion_sen(6, 0, 10, 0)

plt.title("mi funcion sen")
plt.plot(tt,xx)
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.show()    
```
