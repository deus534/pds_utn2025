### Toolbox
Funciones, imports, constantes y todo lo necesario para realizar las pruebas

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
from numpy.fft import fft
from numpy.fft import fftfreq

N = 1000
fs = 1000

def mi_funcion_sen(vmax = 1, dc = 0, ff = 1, ph = 0, nn = N, fs = fs):
    t = nn/fs
    tt = np.arange(0,t,1/fs)
    xx = vmax*np.sin(2*np.pi*ff*tt + ph) + dc
    return tt, xx

def mi_funcion_DFT( xx ):
    N = len(xx)
    n = np.arange(N)
    yy = np.zeros(N,dtype='complex128')
    for k in range(N):
        yy[k] = np.sum(xx * np.exp(-1j*2*np.pi*k*n/N))
    return yy
def mi_funcion_cuadrada(vmax=1, dc=0, ff=1, ph=0, nn=N, fs=fs):
    #me falta la parte de cuando tiene un desfase, que no se me ocurre por ahora
    tt = np.arange(0,nn/fs,1/fs)
    xx = np.zeros(fs)
    for i in range(fs):
        xx[i] = (vmax if (np.sin(2*np.pi*ff*i/fs)>=0) else -vmax ) + dc
    return tt, xx

#Esta funcion graficar solo es para realizarla aca para simplificar las cosas
def graficar(tt, xx, title='grafica'):
    t0dft = time()
    ff = mi_funcion_DFT(xx) 
    tfdft = time()
    
    t0fft = time()
    ffn = fft(xx)
    tffft = time()
    
    freqs = fftfreq(tt.shape[-1],d=1/fs)
    #-----------------------#
    tf = np.arange(N/2)*fs/N
    modulo = np.abs(ff[:N//2])
    fase = np.angle(ff[:N//2])
    
    tfn = freqs[:N//2]
    modulofft = np.abs(ffn)[:N//2]
    fasefft = np.angle(ffn)[:N//2]
    #-----------------------#
    print( f'------------------------DFT----------------------------' )
    print( f'Tamaño de la señal xx: {len(xx)}' )
    print( f'Tamaño del tiempo de la señal xx: {len(tt)}' )
    print( f'Tamaño las frecuencias de la DFT: {len(tf)}')
    print( f'Valor pico de la frecuencia alta: {np.max(modulo)}' )
    print( f'Frecuencia mas alta: {np.argmax(modulo)}' )
    print( f'------------------------FFT----------------------------' )
    print( f'Tamaño de la fft: {len(ffn)}' )
    print( f'Tamaño de las frecuencias de la fft: {len(freqs)}' )
    print( f'Valor pico de la frecuencia alta: {np.max(modulofft)}' )
    print( f'Frecuencia mas alta: {np.argmax(modulofft)}' )
    print( f'----------------------------------------------------' )
    print( f'Tiempo DFT: {tfdft-t0dft}' )
    print( f'Tiempo FFT: {tffft-t0fft}' )
    print( f'diferencia de tiempo {(tfdft-t0dft)-(tffft-t0fft)}' ) 
    #-----------------------#
    fig = plt.figure(figsize=(12,6))
    fig.suptitle(title)
    
    plt.subplot(2,2,1)
    plt.title(f'Modulo de ff(DFT)')
    plt.plot(tf, modulo)
    plt.xlabel('Hz')
    plt.ylabel('| ff |')
    
    plt.subplot(2,2,2)
    plt.title(f'fase de ff(DFT)')
    plt.plot(tf, fase)
    plt.xlabel('Hz')
    plt.ylabel('Rad')
    
    plt.subplot(2,2,3)
    plt.title(f'Modulo de ffn(fft)')
    plt.plot(tfn,modulofft)
    plt.xlabel('Hz')
    plt.ylabel('| ff |')
    
    plt.subplot(2,2,4)
    plt.title(f'Fase de ffn(fft)')
    plt.plot(tfn,fasefft)
    plt.xlabel('Hz')
    plt.ylabel('Rad')
    
    plt.tight_layout()
    plt.show()        
```

### Practicas
En esta seccion se ven las practicas que se realizaran de la TS, siempre buscando obtener todas las incongruencias que se llega a tener o llegar a tener siempre los mayores ejemplos.

#### Pruebas con señales sinusoidales
En esta seccion realizo algunas pruebas con señales sinusoidales viendo que es lo que cambia cuando vario la frecuencia como es que varia si cambia los otros distintos parametros, como el Vmax el offset y tambien la ph si esque llega a cambiar en algo la señal de la DFT

```python
freq = 100
tt, xx = mi_funcion_sen(1,0,freq)

graficar(tt,xx, f'seno de {freq} Hz')
```

```python
Vmax = 1
Dc = 1
desfase = 0
tt, xx = mi_funcion_sen(Vmax, Dc, freq, desfase)

graficar(tt,xx, f'seno de {freq} Hz y DC {Dc}')
```

```python
freq = 10
tt, xx = mi_funcion_cuadrada(1,0,freq)

graficar(tt,xx,f'cuadrada de {freq} Hz')
```
