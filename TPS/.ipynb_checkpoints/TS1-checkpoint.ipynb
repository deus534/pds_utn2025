# Primer ejercicio Semanal

## Desarrollo de la funcion

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
fs = 1000

def mi_funcion_sen(vmax = 1, dc = 0, ff = 1, ph = 0, nn = N, fs = fs):
    t = nn/fs
    tt = np.arange(0,t,1/fs)
    xx = vmax*np.sin(2*np.pi*ff*tt + ph) + dc

    return tt, xx
```

### Pruebas

#### Pruebas basicas
Simplemente variando la frecuencia la amplitud la fase y el componente de DC

```python
tt0, xx0 = mi_funcion_sen(1, 0, 1, 0)
tt1, xx1 = mi_funcion_sen(5, 0, 5, 0)
tt2, xx2 = mi_funcion_sen(1, 1, 10, 0)
tt3, xx3 = mi_funcion_sen(1, 0, 10, 0)

plt.figure()
plt.subplot(2,2,1)
plt.title('seno de 1 Hz')
plt.plot(tt0,xx0)

plt.subplot(2,2,2)
plt.title('seno de 10 Hz')
plt.plot(tt1, xx1)

plt.subplot(2,2,3)
plt.title('seno de 5 Hz')
plt.plot(tt2, xx2)

plt.subplot(2,2,4)
plt.title('seno de 5 Hz')
plt.plot(tt3, xx3)


plt.tight_layout()
plt.show()
```

#### Pruebas medias
En estas pruebas se va tomando frecuencias cercanas a la frecuencia de nyquist o incluso sobrepasando la frecuencia de nyquist viendo si esque se llega a observar lo mismo que en lo anterio o es que llega a cambiar tambien verificando si esque en alguna frecuencia de alguna forma llega a ser normal

```python
tt0, xx0 = mi_funcion_sen(1,0,500,0) #justo en la frecuencia de nyquist
tt1, xx1 = mi_funcion_sen(1,0,499,0)
tt2, xx2 = mi_funcion_sen(1,0,400,0)
tt3, xx3 = mi_funcion_sen(1,0,250,0)
tt4, xx4 = mi_funcion_sen(1,0,260,0)
tt5, xx5 = mi_funcion_sen(1,0,160,0)

plt.figure(figsize=[10,6])

plt.subplot(2,3,1)
plt.title('sen de 500 hz')
plt.plot(tt0, xx0)

plt.subplot(2,3,2)
plt.title('sen de 499 Hz')
plt.plot(tt1,xx1)
plt.xlim([0,0.2])

plt.subplot(2,3,3)
plt.title('sen de 400 hz')
plt.plot(tt2,xx2)
plt.xlim([0,0.1])

plt.subplot(2,3,4)
plt.title('sen de 250 Hz')
plt.plot(tt3, xx3)
plt.xlim([0,0.1])

plt.subplot(2,3,5)
plt.title('sen de 260 Hz')
plt.plot(tt4, xx4)
plt.xlim([0,0.1])

plt.subplot(2,3,6)
plt.title('sen de 160 Hz')
plt.plot(tt5, xx5)
plt.xlim([0,0.1])
plt.tight_layout()
plt.show()
```

#### Conclusiones
Bueno viendo las pruebas realizadas que se ven en si, se puede llegar a observar que para frecuencias bajas esta no llega a afectar para nada, es decir que la frecuencia funciona de maravilla, pero bien llegando a la frecuencia de nyquist, esta empieza a fallar y se llega a observar cosas raras que no es un seno.

Por ejemplo, en la figura del "sen de 500 hz" que es justo en la frecuencia de nyquist esta no se ve ni un poco a un seno y ademas de que la amplitud del "supuesto seno" llega a bajar drasticamente, podriamos decir incluso que llega a cero.

Ahora para frecuencias que estan entre 251 a 499 hz se llegan a observar figuras parecidas al seno, pero en realidad se parece a como que un seno dentro de un seno y dependiendo de la frecuencia se llega a denotar mas o menos este fenomeno.

Finalizando cuando se llega a la frecuencia de 250 hz para abajo el seno sigue sin funcionar del todo bien, no entiendo porque, pense que el seno se llegaba a estabilizar pero en realidad se llega a distorsionar un poco, puedo llegar a pensar es con la resolucion que tiene python lo que causa un poco la distorsion, porque segun nyquist el seno deberia de formarse hasta los 250 hz sin ningun tipo de distorsion pero no se si sera cierto.
