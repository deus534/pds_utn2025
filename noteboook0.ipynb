```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

#Definición de la función IMPULSO UNITARIO APROXIMADO delta(t)
delta = lambda t,Ts: np.piecewise(t, (t>=0) & (t<Ts), [1/Ts,0])

#Definición de la función ESCALÓN UNITARIO u(t)
u = lambda t: np.piecewise(t, t>=0, [1,0])

#Definición de la función RAMPA UNITARIA rho(t)
rho = lambda t: t*u(t)

#FUNCIÓN TFTC (TRANSFORMADA DE FOURIER PARA SEÑALES CONTINUAS - ANÁLISIS)
def TFTC(t, ft, wmax, dw):
    '''
    Calculates the Continuous Fourier Transform

    Parameters:
    t (ndarray):  time vector used to define ft
    ft (ndarray): signal to calculate the STF coefficients
    wmax (value): maximun w value where FT is calculated
    dw: (value):  frequency interval

    Returns:
    Xw (ndarray): FT Spectrum
    w  (ndarray): FT frequency axis

    Example1:
    t = np.arange(0, 10, 0.001)
    x = exp(-2*t)
    Xw,w = TFTC(t, x, 20, 0.001)
    '''
    dt = t[1] - t[0]

    w=np.arange(-wmax,wmax,dw)
    Xw= np.zeros(len(w),dtype=np.complex128);
    for n in range(len(w)):
        Xw[n] = sum(ft*np.exp(-1j*w[n]*t)) * dt

    return Xw,w

#FUNCIÓN ITFTC (TRANSFORMADA INVERSA DE FOURIER PARA SEÑALES CONTINUAS - SÍNTESIS)
def ITFTC(t, Xw, w):
    '''
    Recreate a signal based on its TFTC coefficients

    Receives:
    t (ndarray): time vector used to recreate the signal
    Xw (ndarray): Spectrum from TFTC
    w (ndarray): frequency axis where Xw was calculated

    Returns:
    ft (ndarray): synthesized signal

    Example:
    t = np.arange(0, 10, 0.001)
    x = np.exp(-2*t)
    Xw,w = TFTC(t, ft, 20, 0.001)
    tsyn = np.arange(0, 20, 0.001)
    xsyn = ITFTC(tsyn, Xw, w)
    '''

    ft1 = np.zeros(len(t),dtype=np.complex128)

    for n in range(len(w)):
        ft1 = ft1 + Xw[n] * np.exp(1j*w[n]*t)

    ft1 = ft1.real

    return ft1
```

```python

#obtencion de las señales
Ts = 0.001
t = np.arange(-1,1,Ts)
x = np.exp(-100*t)*u(t)  #entrada
h = np.exp(-100*t)*u(t)  #respesta impulsional

#realizo la convolucion para obtener la salida
y = np.convolve(x,h,mode='same')*Ts

#obtengo el espectro de la convolucion(y)
wmax=350
dw=0.1
Y,w = TFTC(t,y,wmax,dw)

#obtengo la autocorrelacion a partir del Y(w)^2
Ryy = ITFTC(t, np.abs(Y)**2, w)

#visualizacion
plt.subplot(311)
plt.title('seniales')
plt.plot(t,x,t,h)
plt.xlim([-0.25,0.25])
plt.tight_layout()

plt.subplot(312)
plt.title('Autocorrelacion')
plt.plot(t,Ryy)
plt.xlim([-0.25,0.25])
plt.tight_layout()

#densidad espectral de energia
plt.subplot(313)
plt.title('Espectro')
plt.plot(w,np.abs(Y)**2, label='Y(w)^2')
plt.legend()
plt.tight_layout()
```
