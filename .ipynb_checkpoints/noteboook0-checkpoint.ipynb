```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,1,0.001)
x = 3*np.sin(2*np.pi*t*10)

plt.title("figura")
plt.plot(t,x)
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.show()
```
